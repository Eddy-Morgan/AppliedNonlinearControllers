```python
import scipy as si
import numpy as np
from scipy.integrate import solve_ivp
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def dSdt(t,S):
    #print(t,S)
    x1,x2 = S
    c1 = 2
    c2 = 3
    C1 = 5
    C2 = 8
    a = 5
    k = 15
    pp = 0.4
    xdot = np.zeros([2,1])
    s = a - (a*x1) - x2
    u = (-a*x2) - (C1*(x2**2)) - (C2*x1) + (k*(s**pp))
    xdot[0] = x2
    xdot[1] = (c1*(x2**2)) + (c2*x1) + u
    return [
        xdot[0],
        xdot[1],
    ]
```

```python
t_end = 10
t_span = (0, t_end)
y0 = [0,0.2]
ans = solve_ivp(dSdt, t_span, y0, method='LSODA')
ans.message
```

```python
x1,x2,t_int = ans.y[0] , ans.y[1], ans.t
plt.rcParams["figure.figsize"] = (20,15)
plt.rcParams.update({'font.size': 22})
fig, (ax1, ax2) = plt.subplots(2)
ax1.plot(t_int, x1, label = "Theta0",linewidth=3)
ax1.set_xlabel("Time (s)")
ax1.set_ylabel("x1")
ax1.set_title("Plot of x1 reference tracking")

ax2.plot(t_int, x2, label = "dTheta0", linewidth=3)
ax2.set_xlabel("Time (s)")
ax2.set_ylabel("x2")
ax2.set_title("Plot of x2 reference tracking")
```
